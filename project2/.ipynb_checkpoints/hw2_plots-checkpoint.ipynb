```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 4]
y_64 = [.240, .200, .151]
y_128 = [.238, .184, .098]

fig, ax = plt.subplots(figsize=(9,5))
ax.set_ylabel('Miss Rate')
ax.set_xlabel('Associativity')
ax.set_title('64 set split vs 128 set unified cache performance')
plt.plot(x, y_64, label='Effective Combined Cache Miss Rate')
plt.plot(x, y_128, label='Unified Cache Miss Rate')
plt.legend()
fig.tight_layout()
plt.show()
```

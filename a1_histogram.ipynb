```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load, store, uncond branch, cond branch, int, float
labels = ['load', 'store', 'uncond branch', 'cond branch', 'int', 'float']
alpha_math = [17.16, 10.47, 3.83, 10.79, 55.62, 2.00]
alpha_fmath = [17.75, 12.76, 4.57, 10.95, 53.40, 0.46]
alpha_llong = [17.85, 14.94, 5.49, 12.03, 49.41, 0.11]
alpha_printf = [18.40, 11.09, 4.60, 11.06, 54.64, 0.09]

pisa_math = [14.30, 10.84, 4.29, 12.69, 56.88, 0.99]
pisa_fmath = [12.98, 15.41, 4.43, 13.45, 53.58, 0.14]
pisa_llong = [10.09, 21.63, 4.75, 12.23, 51.26, 0.00]
pisa_printf = [14.30, 9.19, 5.83, 14.37, 56.31, 0.02]
```

```python
def plot_bargraph(labels, alpha_x, pisa_x, test_str):
    x = np.arange(len(labels))  # the label locations
    width = .35  # the width of the bars

    fig, ax = plt.subplots(figsize=(9,5))
    rects1 = ax.bar(x - width/2, alpha_x, width, label='Alpha')
    rects2 = ax.bar(x + width/2, pisa_x, width, label='PISA')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Percentage of instructions')
    ax.set_title('Percentage of Instructions by ISA for ' + test_str)
    ax.set_xticks(x, labels)
    ax.legend()

    ax.bar_label(rects1, padding=3)
    ax.bar_label(rects2, padding=3)

    fig.tight_layout()

    plt.show()
```

```python
plot_bargraph(labels, alpha_math, pisa_math, 'test-math')
```

```python
plot_bargraph(labels, alpha_fmath, pisa_fmath, 'test-fmath')
```

```python
plot_bargraph(labels, alpha_llong, pisa_llong, 'test-llong')
```

```python
plot_bargraph(labels, alpha_printf, pisa_printf, 'test-printf')
```
